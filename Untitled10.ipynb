```python
# Task 1 – User Login Check

username = "admin"
password = "1234"

# Predefined correct credentials
correct_username = "admin"
correct_password = "1234"

# Check if both username and password match
if username == correct_username and password == correct_password:
    print("Login Successful")
else:
    print("Invalid Credentials")
```

```python
# Task 2 – Pass / Fail Analyzer

marks = [45, 78, 90, 33, 60]

pass_count = 0
fail_count = 0

# A student passes if marks >= 50
for m in marks:
    if m >= 50:
        pass_count += 1
    else:
        fail_count += 1

print(f"Total Students : {len(marks)}")
print(f"Passed         : {pass_count}")
print(f"Failed         : {fail_count}")
```

```python
# Task 3 – Simple Data Cleaner

names = [" Alice ", "bob", " CHARLIE "]

cleaned_names = []
for name in names:
    cleaned = name.strip().lower()   # remove spaces + lowercase
    cleaned_names.append(cleaned)

print("Cleaned names:", cleaned_names)
```

```python
# Task 4 – Message Length Analyzer

messages = ["Hi", "Welcome to the platform", "OK"]

for msg in messages:
    length = len(msg)
    if length > 10:
        flag = "  ⚑ LONG MESSAGE"
    else:
        flag = ""
    print(f'"{msg}"  ->  Length: {length}{flag}')
```

```python
# Task 5 – Error Message Detector

logs = ["INFO", "ERROR", "WARNING", "ERROR"]

error_count = 0

for log in logs:
    if log == "ERROR":
        error_count += 1

print(f"Total log entries : {len(logs)}")
print(f"ERROR count       : {error_count}")
```
