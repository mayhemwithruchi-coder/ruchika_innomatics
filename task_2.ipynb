```python
numbers = [45, 22, 89, 10, 66]
numbers = [45, 22, 89, 10, 66]
max_val = numbers[0]
min_val = numbers[0]
for num in numbers:
    if num > max_val:
        max_val = num
    if num < min_val:
        min_val = num
print('List:', numbers)
print('Max:', max_val)
print('Min:', min_val)
```
